```python
# import libraries
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data
fashion_mnist = keras.datasets.fashion_mnist

# split the data
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# check the data shape
print(train_images.shape, train_labels.shape)
print(test_images.shape, test_labels.shape)
```

```python
# check the data
train_images[0]
```

```python
# check the labels
train_labels[:10]
```

```python
# check the class names
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# show an image
plt.figure()
plt.imshow(train_images[1])
plt.colorbar()
plt.grid(True)
plt.show()
```

```python
# preprocess the data -> making numbers smaller in order to reduce the size of the data
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# build the model -> the neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), # here we flatten the images
    keras.layers.Dense(128, activation='relu'), # here we add a dense layer
    keras.layers.Dense(10, activation='softmax') # here we add a dense layer
])
```

```python
# compile the model
model.compile(optimizer='adam', # here we use the adam optimizer -> it is a gradient descent algorithm
              loss='sparse_categorical_crossentropy', # here we use the sparse categorical crossentropy loss
              metrics=['accuracy']) # here we use the accuracy metric
```

```python
# train the model
model.fit(train_images, train_labels, epochs=10)
```

```python
# evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

```python
for i in range(20):
    # predict the model
    predictions = model.predict(test_images)
    print(class_names[np.argmax(predictions[i])])

    # show the image
    plt.figure()
    plt.imshow(test_images[i])
    plt.colorbar()
    plt.grid(False)
    plt.show()
```
